```python
# -*- coding:utf-8 -*-
import nltk

nltk.download('punkt')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
```

```python
sentence = "Prime Minister Boris Johnson had previously said the UK would leave by 31 October."
```

```python
tokens = nltk.word_tokenize(sentence)
print(tokens)
```

```python
tagged = nltk.pos_tag(tokens)
print(tagged)
```

```python
entities = nltk.chunk.ne_chunk(tagged)
print(entities)
```

### spaCy 패키지에서 NER(Named Entity Recognition)을 지원하는 언어

image.png

```python
import spacy
nlp = spacy.load("en_core_web_sm")
text = "Apple is looking at buying U.K. startup for $1 billion"
doc = nlp(text)
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
# 대구대학교는 1956년 고(故) 이영식 박사에 의해 설립되었습니다.
# 한국 전쟁 이후, 전쟁으로 인해 상처받은 사람들에게 교육의 기회를 제공하고자 하는 취지에서 시작되었습니다.

text = """Daegu University was founded in 1956 by Dr. Lee Young-Sik.
After the Korean War, it was established with the purpose of providing educational opportunities to those affected by the war.
"""

doc = nlp(text)
for ent in doc.ents:
    print(ent.text, ent.label_)
```
